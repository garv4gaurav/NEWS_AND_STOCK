```
#http://archive.indianexpress.com/
import re
import requests
from time import time
from openpyxl import Workbook
from bs4 import BeautifulSoup

starttime = time()
IE_ARCH_URL = "http://archive.indianexpress.com/"
response = requests.get(IE_ARCH_URL)
print(response)


SOUP = BeautifulSoup(response.content, 'lxml')
urls = []

for a_tag in SOUP.find_all('a'):
    if ('archive' in a_tag.attrs['href']) and (str(a_tag.attrs['href']) != 'http://archive.indianexpress.com/'):
        #print('http://archive.indianexpress.com'+str(a_tag.attrs['href']))
        urls.append('http://archive.indianexpress.com'+str(a_tag.attrs['href']))

print(len(urls))


workbook = Workbook()
sheet = workbook.active
sheet['A1'] = 'Base URL with Date Month Year'
sheet['B1'] = 'Business URLS'
sheet['C1'] = 'Date'
sheet['D1'] = 'Content'


def get_business_urls(url):
    urls = []
    response = requests.get(url)
    if response.status_code == 200:
        SOUP = BeautifulSoup(response.content, 'lxml')
        for tag in SOUP.find_all('h4', class_='arch1'):
            if tag.contents[0] == 'Business': 
                page_link_list = tag.next_sibling.find_all('a')
                for a_tag in page_link_list:
                    #print(str(a_tag.attrs['href']))
                    urls.append(str(a_tag.attrs['href']))
    else:
        urls.append('No links found.')
    return urls



def get_paragraph(url):
    paragraph = ''
    date = ''
    response = requests.get(url)
    if response.status_code == 200:
        SOUP = BeautifulSoup(response.content, 'lxml')
        para_list = SOUP.find_all('p')
        for para in para_list:
            paragraph = paragraph + str(para.text)
        div_tag = SOUP.find_all('div', class_ = 'story-date')
        for a in div_tag:
            #print(a.text.strip())
            date = a.text.strip()
    else:
        urls.append('No data found.')
    return paragraph, date


count = len(urls)


for i in range(count):
    if i < 2000:
        continue
    else:
        if i%200 == 0:
            workbook.save('/content/drive/My Drive/'+'Data_'+ str(i) +'.xlsx')
            workbook = Workbook()

            sheet = workbook.active
            sheet['A1'] = 'Base URL with Date Month Year'
            sheet['B1'] = 'Business URLS'
            sheet['C1'] = 'Date'
            sheet['D1'] = 'Content'

        business_links = get_business_urls(urls[i])
        for j in range(len(business_links)):
            sheet['A'+str(i+j+2)] = urls[i]
            sheet['B'+str(i+j+2)] = business_links[j]
            if re.search("^/oldStory/", business_links[j]):
                new_business_url = 'http://archive.indianexpress.com' + business_links[j]
                para, date = get_paragraph(new_business_url)
            else:
                para, date = get_paragraph(business_links[j])
            #print(para, date)
            sheet['C'+str(i+j+2)] = date
            sheet['D'+str(i+j+2)] = para
        print(time()-starttime, i)


workbook.save('/content/drive/My Drive/'+'Data'+str(count+2)+'.xlsx')


```
